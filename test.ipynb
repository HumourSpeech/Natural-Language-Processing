```python
import speech_recognition as sr
from googletrans import Translator

def speech_to_text(lang_code):
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print(f"Listening... (Language: {lang_code})")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio, language=lang_code)
        print(f"Recognized Text: {text}")
        return text
    except sr.UnknownValueError:
        print("Sorry, could not understand the audio.")
        return None
    except sr.RequestError:
        print("Request failed. Check your internet connection.")
        return None

def translate_text(text):
    translator = Translator()
    translated = translator.translate(text, dest="en")
    print(f"Translated Text: {translated.text}")
    return translated.text

if __name__ == "__main__":
    language_codes = {"english": "en", "hindi": "hi", "odia": "or"}

    print("Choose a language: English, Hindi, Odia")
    lang_choice = input("Enter language: ").strip().lower()

    if lang_choice in language_codes:
        lang_code = language_codes[lang_choice]
        spoken_text = speech_to_text(lang_code)

        if spoken_text and lang_choice != "english":
            translated_text = translate_text(spoken_text)
        else:
            translated_text = spoken_text

        if translated_text:
            with open("bank_form_data.txt", "a", encoding="utf-8") as file:
                file.write(translated_text + "\n")
                print("Data saved successfully!")
    else:
        print("Invalid language selection.")
```

```python
import speech_recognition as sr
from googletrans import Translator

def speech_to_text(lang_code):
    recognizer = sr.Recognizer()
    
    with sr.Microphone() as source:
        print(f"Listening... (Language: {lang_code})")
        recognizer.adjust_for_ambient_noise(source)
        
        try:
            # First 30 seconds of speech input
            print("You have 30 seconds to speak...")
            audio = recognizer.listen(source, timeout=30)
            text = recognizer.recognize_google(audio, language=lang_code)
            
            print(f"Recognized Text: {text}")
            
            # Asking for an extension if needed
            print("\nDo you need more time? (Say 'yes' to continue, anything else to stop)")
            audio_extension = recognizer.listen(source, timeout=5)
            response = recognizer.recognize_google(audio_extension, language="en").lower()
            
            if "yes" in response:
                print("You have another 30 seconds...")
                audio_extended = recognizer.listen(source, timeout=30)
                text_extended = recognizer.recognize_google(audio_extended, language=lang_code)
                text += " " + text_extended  # Combine the first and extended texts
            
            return text

        except sr.UnknownValueError:
            print("Sorry, could not understand the audio.")
        except sr.RequestError:
            print("Request failed. Check your internet connection.")
        except sr.WaitTimeoutError:
            print("No speech detected within the time limit.")
    
    return None

def translate_text(text):
    translator = Translator()
    translated = translator.translate(text, dest="en")
    print(f"Translated Text: {translated.text}")
    return translated.text

if __name__ == "__main__":
    language_codes = {"english": "en", "hindi": "hi", "odia": "or"}

    print("Choose a language: English, Hindi, Odia")
    lang_choice = input("Enter language: ").strip().lower()

    if lang_choice in language_codes:
        lang_code = language_codes[lang_choice]
        spoken_text = speech_to_text(lang_code)

        if spoken_text and lang_choice != "english":
            translated_text = translate_text(spoken_text)
        else:
            translated_text = spoken_text

        if translated_text:
            with open("bank_form_data.txt", "a", encoding="utf-8") as file:
                file.write(translated_text + "\n")
                print("Data saved successfully!")
    else:
        print("Invalid language selection.")
```
